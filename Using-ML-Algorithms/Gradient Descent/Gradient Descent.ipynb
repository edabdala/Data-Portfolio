```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use("ggplot")
```

```python
# random function to minimize
function = lambda x: (x**2)-(2*x)
```

$y = x^2 + 2$

```python
x = np.linspace(-1, 3, 500)
plt.plot(x, function(x))
plt.show()
```

minimum is at $x = -1$

```python
def derivative(x): 
    deriv = 2*x + 2
    return deriv
```

```python
def stepfunc(start, x_prev, precision, lrate):    
    x_list, y_list = [start], [function(start)]
    while abs(start - x_prev) > precision:
        
        x_prev = start
        
        d_x = -derivative(x_prev)
        # tells us how much we will descend, learning rate set arbitrarily
        start = x_prev + (lrate * d_x)
        
        x_list.append(start)
        
        y_list.append(function(start))
    print ("Local minimum occurs at: "+ str(start))
    print ("Number of steps: " + str(len(x_list)))
    
    
    plt.subplot(1,2,2)
    plt.scatter(x_list,y_list,c="green")
    plt.plot(x_list,y_list,c="green")
    plt.plot(x,function(x), c="yellow")
    plt.title("Gradient descent")
    plt.show()
```

```python
stepfunc(10,1, .001, .05)
```
